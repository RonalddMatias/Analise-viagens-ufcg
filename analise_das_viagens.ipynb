# Análise de gastos das viajens da UFCG

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/Guardians-DSC/dados/main/orcamento/dados_portal_tranparencia_ufcg_2019_2023.csv')
```

```python
df
```

```python
df.head()
```

## 1 Passo: Selecionar as linhas importantes do dataframe

-> Vendo os elementos unicos da coluna elemento_de_despesa

```python
df['elemento_de_despesa'].unique()
```

Iremos fazer uma filtragem da coluna 'elemento_de_despesa' e selecionar apenas aquelas do tipo '33 - Passagens e Despesas com Locomoção'

```python
df = df.loc[df['elemento_de_despesa'] == '33 - Passagens e Despesas com Locomoção']
df
```

## 2 Passo: Tratando os dados nulos

```python
df.isnull().sum()
```

percebemos que a coluna "unnamed:_22" possuem todos null, então vamos retira-la do nosso dataframe


```python
df = df.drop('unnamed:_22', axis = 1)
```

```python
df.columns
```

```python
df['valor']
```

## 3 Passo: Ocorrencia das viagens de acordo com os anos

```python
#df.groupby(df.your_date_column.dt.year)['values_column'].sum()
df['data'] = pd.to_datetime(df['data']) # transformando em um dataframe
ocorrencia_por_ano = df.groupby(df['data'].dt.year)['elemento_de_despesa'].count().sort_values()
ocorrencia_por_ano
```

```python
ocorrencia_por_ano.plot(kind='barh', figsize=(10,10), xlabel = 'ocorrências', ylabel = 'anos', color = 'purple')
# COLOCAR TITULO
```

Ao analisar a ocorrencia de gastos por ano, percebemos que o ano de 2019 foi o que mais gastou com viajens nos ultimos 4 anos.

2020 e 2021, anos pandemicos, foram os que tiverem a menor taxa de gastos com viajens

2023 aparenta estar subindo os gastos com viajens.

```python
# porcentagem de cada ano em relação ao total
porcentagem_por_ano = ocorrencia_por_ano / ocorrencia_por_ano.sum() * 100
porcentagem_por_ano = porcentagem_por_ano.round(2)
porcentagem_por_ano
```

```python
porcentagem_por_ano.plot(kind='pie', autopct='%1.1f%%', ylabel = '') # autopct é usado para mostrar a pocentagem no grafico
plt.axis('equal')
plt.title('Porcentagem de ocorrências por ano', y = 1.1)
plt.show()
```

# 4 Passo: analisaremos as ocorrências de viagens por meses dos anos

## 2019

```python
df['data'] = pd.to_datetime(df['data'])
df_2019 = df[df['data'].dt.year == 2019]
ocorrencias_por_mes = df_2019.groupby(df_2019['data'].dt.month)['elemento_de_despesa'].count().sort_index()
ocorrencias_por_mes
```

Renomeando os index

```python
meses = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
ocorrencias_por_mes.index = meses
ocorrencias_por_mes
```

```python
ocorrencias_por_mes.plot(kind='bar', rot=0)
plt.xlabel('Mês')
plt.ylabel('Ocorrência')
plt.title('Ocorrências por mês em 2019')
plt.show()
```

-> Ordenei os index para a identificação dos meses ficarem mais faceis

```python
porcentagem_por_mes_2019 = ocorrencias_por_mes / ocorrencias_por_mes.sum() * 100
porcentagem_por_mes_2019 = porcentagem_por_mes_2019.round(2)
porcentagem_por_mes_2019.index = meses
porcentagem_por_mes_2019
```

```python
porcentagem_por_mes_2019.plot(kind= 'pie',autopct='%1.1f%%', ylabel = '')
```
